```python
import numpy as np
import pandas as pd
from sklearn.metrics import root_mean_squared_error  
from tabulate import tabulate
```

```python
"""
1. Crea dos vectores con 10 valores aleatorios int entre 30 y 60 (vector 1) y con 10 valores float del 0.1 al 9.9 (vector 2) 
2. El primer vector se corresponderá con el valor real y el segundo vector con el predicho por el modelo
3. Calcula el RMSE con la función de scikit y con la fórmula tradicional.
4. Imprime el resultado
5. ¿Hay diferencia entre los resultados?
"""
v_real = np.random.randint(29,60,10)
v_predicho = np.random.random((10)) 
v_predicho *= 10

rmse_scikit = root_mean_squared_error(v_real,v_predicho)
print(f"RMSE con la función de scikit: {rmse_scikit:.2f}")

errores = (v_real - v_predicho) ** 2
mse = np.mean(errores)
rmse_formula = np.sqrt(mse)
print(f"RMSE con la fórmula tradicional: {rmse_formula:.2f}")

```

```python
df_shipment = pd.read_csv("ShipmentLine_v3.csv")
orderId = np.array(df_shipment["order.orderIdentifier"])
value = np.array(df_shipment["value"])

df_shipment['orderIdNormalizado'] = (orderId - orderId.mean()) / orderId.std()
df_shipment['valueNormalizado'] = (value - value.mean()) / value.std()
df_shipment['desviacion_orderNormalizado'] = df_shipment['orderIdNormalizado'].std()
df_shipment['desviacion_valueNormalizado'] = df_shipment['valueNormalizado'].std()

df_shipment
```

```python

# DataFrames de ejemplo
df1 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Nombre': ['Ana', 'Luis', 'Pedro', 'Marta']
})

df2 = pd.DataFrame({
    'ID': [3, 4, 5],
    'Salario': [25000, 45000, 35000]
})

# Fusión de ambos DataFrames por la columna 'ID' (común)
df_merged = pd.merge(df1, df2, how='inner', on='ID')
print(df_merged)

# Fusión externa para mantener todos los datos
df_merged_outer = pd.merge(df1, df2, how='outer', on='ID')
print(df_merged_outer)

# Fusión utilizando diferentes nombres de columnas
df1 = pd.DataFrame({
    'ID_cliente': [1, 2, 3, 4],
    'Nombre': ['Ana', 'Luis', 'Pedro', 'Marta']
})

df2 = pd.DataFrame({
    'ID_usuario': [3, 4, 5],
    'Salario': [25000, 45000, 35000]
})

# Fusión utilizando las columnas 'ID_cliente' de df1 y 'ID_usuario' de df2
df_merged_diff_columns = pd.merge(df1, df2, how='inner', left_on='ID_cliente', right_on='ID_usuario')
print(df_merged_diff_columns)
```

```python
# Crear un DataFrame de ejemplo
data = {
    'Departamento': ['Ventas', 'Ventas', 'IT', 'IT', 'RRHH', 'RRHH', 'Ventas'],
    'Empleado': ['Ana', 'Luis', 'Pedro', 'Marta', 'Sofía', 'Carlos', 'David'],
    'Salario': [30000, 50000, 60000, 55000, 40000, 42000, 48000],
    'Edad': [25, 30, 40, 35, 28, 45, 32],
    'sexo': ["Hombre","Mujer","Hombre","Mujer","Hombre","Hombre","Mujer"]
}

df = pd.DataFrame(data)

# Agrupar por varias columnas (Departamento y Edad agrupada en rangos)
df['Grupo_Edad'] = pd.cut(df['Edad'], bins=[20, 30, 40, 50], labels=['20-30', '30-40', '40-50'])
grupo_edad_salario = df.groupby(['Departamento', 'Grupo_Edad'])['Salario'].mean()
print("Salario promedio por departamento y grupo de edad:\n", grupo_edad_salario, "\n")

"""Crea una columna que especifique el sexo de los empleados. A continuación, agrupa los datos por Departamentos y además por sexo, especifica etiquetas "Hombre y "Mujer para cada grupo """

grupo_departamento_sexo = df.groupby(['Departamento', 'sexo']).count()
print("depasexo", grupo_departamento_sexo)

```
